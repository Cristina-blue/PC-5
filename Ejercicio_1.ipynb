```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

file_path = '/workspaces/PC-5/airbnb.csv'
df = pd.read_csv(file_path)

print("Primeras filas del dataset:")
print(df.head(), "\n")

print(f"El dataset tiene {df.shape[0]} filas y {df.shape[1]} columnas.\n")

print("Nombres de las columnas:")
print(df.columns.tolist(), "\n")

print("Tipos de datos de cada columna:")
print(df.dtypes, "\n")

print("Resumen estadístico:")
print(df.describe(), "\n")

print("Valores faltantes en cada columna:")
print(df.isnull().sum(), "\n")

print("Valores únicos en columnas clave:")
print(f"Tipos de propiedades: {df['room_type'].nunique()} únicos")
print(f"Barrios: {df['neighborhood'].nunique()} únicos\n")

plt.figure(figsize=(10, 6))
sns.boxplot(x=df['price'])
plt.title('Distribución de Precios en Airbnb')
plt.xlabel('Precio (en euros por noche)')
plt.show()
```
