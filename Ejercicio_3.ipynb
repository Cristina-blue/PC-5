```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

file_path = '/workspaces/PC-5/airbnb.csv'  # Asegrate de que la ruta sea correcta
df = pd.read_csv(file_path)

precio_medio_por_barrio = df.groupby('neighborhood')['price'].mean().reset_index()
precio_medio_por_barrio = precio_medio_por_barrio.sort_values(by='price', ascending=False)

conteo_por_tipo = df['room_type'].value_counts().reset_index()
conteo_por_tipo.columns = ['room_type', 'count']

plt.figure(figsize=(12, 6))
sns.barplot(x='price', y='neighborhood', data=precio_medio_por_barrio, palette='viridis')
plt.title('Precio Medio de Propiedades por Barrio en Lisboa')
plt.xlabel('Precio Medio (en euros)')
plt.ylabel('Barrio')
plt.show()


plt.figure(figsize=(8, 5))
sns.barplot(x='count', y='room_type', data=conteo_por_tipo, palette='pastel')
plt.title('Nmero de Propiedades por Tipo de Habitacin')
plt.xlabel('Nmero de Propiedades')
plt.ylabel('Tipo de Habitacin')
plt.show()
```
